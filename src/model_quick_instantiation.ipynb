# Quick demo of the Instant Policy Model
This is a simple model instantiation to check that the dimensions match and tensors work right with the previous data from the heterogeneous graph representation.

```python
from src.classes.instant_policy_model import InstantPolicyModel

# 1. Recuperamos un grafo de prueba generado en tu etapa 2
graph = load_demo_as_graph(HDF5_PATH, 'demo_0', k_neighbors=5)

# 2. Instanciamos el modelo usando la "metadata" de tu grafo
# La metadata es una tupla con (lista_de_tipos_de_nodos, lista_de_tipos_de_aristas)
model = InstantPolicyModel(metadata=graph.metadata(), node_features=2, hidden_dim=64)

# 3. Simulamos un batch y un timestep aleatorio
# Asumimos que estamos en el paso de difusión t=42
simulated_timestep = torch.tensor([42], dtype=torch.float32)

# IMPORTANTE: Simulamos la inyección de ruido en la acción
# (Esto lo harás rigurosamente en el bucle de entrenamiento)
ruido_simulado = torch.randn_like(graph['action'].x)
graph['action'].x = graph['action'].x + ruido_simulado

# 4. Hacemos un Forward Pass
prediccion = model(graph, simulated_timestep)

print("Forma del tensor de entrada (nodos action):", graph['action'].x.shape)
print("Forma de la predicción de ruido:", prediccion.shape)
```
